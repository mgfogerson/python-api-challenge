```python
#importing dependencies
import csv
import requests
import pandas as pd
import json
from scipy.stats import linregress
import numpy as np
from citipy import citipy
from pprint import pprint
import matplotlib.pyplot as plt
#api key in config
from config import api_key
```

```python
#create randomized lists of cities
#n is the number, 600 cities to be +100 of the project threshold
cities = []
coords = []
csv_cities = []
#create uniform, random latitude/longitude points, and
#use citipi to collect the nearest city by zipped coords
lat = np.random.uniform(low=-90, high=90, size=2000)
lng = np.random.uniform(low=-180, high=180, size=2000)
#store coordinates as a tuple
coords = zip(lat, lng)
#iterate over coords using for loop
for x in coords:
    city = citipy.nearest_city(x[0], x[1]).city_name
    if city not in cities:
        cities.append(city)

#and then have the loop randomize more coordinate pairs

display(len(cities))



```

```python
#use weather API 
url = "http://api.openweathermap.org/data/2.5/weather?"
#build query url
testing_url = url + "appid=" + api_key + "&q=" + "San Diego" + "&units=" + "imperial"
#print to test response and get indexes for the values I need later
```

```python
#check that API is responding
test_response = requests.get(testing_url)
test_json = test_response.json()
pprint({json.dumps(test_json, indent=2,)})
#testing out how to isolating the temp as an example for later
print(test_json['main']['temp'])
      
      
      
```

```python
display(cities)
```

```python
#create lists to append conditions to for each city
city_number = 1
city_name = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
latitude = []
longitude = []
#begin text output as city weather is processed:
print('--------------------------')
print('Start of Weather Records')
print('--------------------------')
#use API in a for loop to iterate through generated cities
for x in cities:
    try:
        weather_url = url + "appid=" + api_key + "&q=" + x + "&units=" + "imperial"
        weather_response = requests.get(weather_url)
        weather_json = weather_response.json()
        city_name.append(weather_json['name'])
        temperature.append(weather_json['main']['temp'])
        humidity.append(weather_json['main']['humidity'])
        cloudiness.append(weather_json['clouds']['all'])
        wind_speed.append(weather_json['wind']['speed'])
        latitude.append(weather_json['coord']['lat'])
        longitude.append(weather_json['coord']['lon'])
        print(f'City {x} processed. {city_number} out of {len(cities)} complete.')
        city_number = city_number + 1
        csv_cities.append(x)
#Realized after some time that I was not getting a key error so much as a notification 
#of incomplete jsons in the API call. Adding KeyError exception fixes issue. 
    except KeyError:
        print(f'City {x} missing data. Skipping {x}. {city_number} out of {len(cities)} complete.')
        city_number = city_number+1
print('--------------------------')
print('Weather Records Complete')
print('--------------------------')

```

```python
#make dataframe out of appended lists
weather_df = pd.DataFrame({
    "City": city_name,
    "Longitude": longitude,
    "Latitude": latitude,
    "Temperature": temperature,
    "Wind Speed": wind_speed,
    "Humidity": humidity,
    "Cloudiness": cloudiness
})
pd.set_option("display.max_rows", None, "display.max_columns", None)
weather_df.to_csv('Global Weather Data.csv', index=False)
weather_df
```

```python
temp_plot = weather_df.plot.scatter(x='Latitude', y="Temperature")
plt.title("Latitude v. Temperature")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Temperature (F)')
display(temp_plot)
plt.savefig('Lat_v_Temp.png')

```

As the latitude approaches the equator at 0, the temperature rises. At the latitudes farther north/south, the temperature dips. Interestingly, there seems to be a flattened out area near the equator, where the rate of temperature increase slows down. 

```python
humid_plot = weather_df.plot.scatter(x='Latitude', y="Humidity")
plt.title("Latitude v. Humidity")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Humidity %')
display(humid_plot)
plt.savefig('Lat_v_Humid.png')
```

There is surprisingly not a great deal of relationship between humidity and latitude. Humidity between 20-100 is found at most latitudes while, between latitudes 15-50 there is a chunk of cities nearer to 0 humidity. Without digging into individual cities, I suspect the dry Sahara/Northern Africa to be a contributor.

```python
cloud_plot = weather_df.plot.scatter(x='Latitude', y="Cloudiness")
plt.title("Latitude v. Cloudiness")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Cloud Cover')
plt.savefig('Lat_v_Cloud.png')
```

Cloud cover is measured in octas, or eighths of the sky covered. There seems to be fairly even distribution of cloud cover, with 6 octa increments emerging as the most frequent.  

```python
windspeed_plot = weather_df.plot.scatter(x='Latitude', y="Wind Speed")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Wind Speed (mph)')
plt.title("Latitude v. Wind Speed ")
plt.savefig('Lat_v_Wind.png')
```

There is a W-shaped curve apparent, where winds are strongest at the extreme north and south as well as at the equator. The northernmost latitudes have the strongest winds recorded.

```python
#use loc to split df along the equator at 0 latitude
northern_df = weather_df.loc[(weather_df)["Latitude"]>0]
southern_df = weather_df.loc[(weather_df)["Latitude"]<0]
northern_df.to_csv('Northern Hemisphere Data.csv', index=False)
southern_df.to_csv('Southern Hemisphere Data.csv', index=False)
display(northern_df)
display(southern_df)
```

```python
x_ntemp = northern_df['Latitude']
y_ntemp = northern_df['Temperature']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_ntemp, y_ntemp)
regress_values = x_ntemp * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_ntemp,regress_values,"r-")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Temperature (F)')
plt.scatter(x_ntemp, y_ntemp, color='k', s=3)
plt.title("Latitude v. Temperature for the Northern Hemisphere")
plt.savefig('Lat_v_Temp_North.png')
plt.show()
```

```python
x_stemp = southern_df['Latitude']
y_stemp = southern_df['Temperature']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_stemp, y_stemp)
regress_values = x_stemp * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_stemp,regress_values,"r-",)
plt.xlabel('Degrees of Latitude')
plt.ylabel('Temperature (F)')
plt.scatter(x_stemp, y_stemp, color='k', s=3)
plt.title("Latitude v. Temperature for the Southern Hemisphere")
plt.savefig('Lat_v_Temp_South.png')
plt.show()
```

The temperature pattern holds true when isolating each hemmisphere. As the city locale approaches the equator at 0, there is an increase in temperature, which drops as the city's location is more north or south of 0. 

```python
x_nhum = northern_df['Latitude']
y_nhum = northern_df['Humidity']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_nhum, y_nhum)
regress_values = x_nhum * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_nhum,regress_values,"r-",label='y={:.2f}x+{:.2f}'.format(slope,intercept))
plt.xlabel('Degrees of Latitude')
plt.ylabel('Humidity %')
plt.scatter(x_nhum, y_nhum, color='k', s=3)
plt.title("Latitude v. Humidity for the Northern Hemisphere")
plt.savefig('Lat_v_Humid_North.png')
plt.show()
```

```python
x_shum = southern_df['Latitude']
y_shum = southern_df['Humidity']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_shum, y_shum)
regress_values = x_shum * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_shum,regress_values,"r-")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Humidity %')
plt.scatter(x_shum, y_shum, color='k', s=3)
plt.title("Latitude v. Humidity for the Southern Hemisphere")
plt.savefig('Lat_v_Humid_South.png')
plt.show()
```

Humidity has a weak correlation with latitude. However, the same pattern from the bi-hemispheric chart was noticed, where 0 humidity is found between 10 and 50 (or -10 and -50) degrees latitude. 

```python
x_nwind = northern_df['Latitude']
y_nwind = northern_df['Wind Speed']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_nwind, y_nwind)
regress_values = x_nwind * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_nwind,regress_values,"r-")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Wind Speed (mph)')
plt.scatter(x_nwind, y_nwind, color='k', s=3)
plt.title("Latitude v. Wind Speed for the Northern Hemisphere")
plt.savefig('Lat_v_Wind_North.png')
plt.show()
```

```python
x_swind = southern_df['Latitude']
y_swind = southern_df['Wind Speed']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_swind, y_swind)
regress_values = x_swind * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_swind,regress_values,"r-")
plt.xlabel('Degrees of Latitude')
plt.ylabel('Wind Speed (mph)')
plt.scatter(x_swind, y_swind, color='k', s=3)
plt.title("Latitude v. Wind Speed for the Southern Hemisphere")
plt.savefig('Lat_v_Wind_South.png')
plt.show()
```

There is a  correlation between wind speed and distance from the equator. It is not as strong as temperature but seemingly stronger than humidity. The most extreme average wind speeds are in the further north or south latitudes, although most of these latitudes have the same low-speed winds as well.
